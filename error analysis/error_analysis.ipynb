### Which tokens were easier to predict? Which of those were in train data?

```python
train_df = pd.read_csv('SEM2012_training_data_with_features.csv')
unique_tokens_train = train_df['lemma'].unique()
```

```python
equal = val_df.loc[val_df['label']!='O']
equal = equal.loc[equal['label']== equal['prediction']]
equal
```

```python
unique_tokens = equal['lemma'].unique()
equal['lemma'].count()

# Create dictionary with unique lemmas as keys and their true positive prediction count as values
token_true_pred = {}
token_true_pred_no_train = {}
token_true_pred_in_train = {}
for item in unique_tokens:
    token_true_pred[item] = 0
    token_true_pred_no_train[item] = 0
    token_true_pred_in_train[item] = 0

# token_true_pred
```

```python

for key, value in token_true_pred.items():
    for index, row in equal.iterrows():
        if row['lemma'] == key:
            token_true_pred[key] += 1
        if row['lemma'] == key and row['lemma']  in unique_tokens_train:
            token_true_pred_in_train[key] += 1
        if row['lemma'] == key and row['lemma'] not in unique_tokens_train:
            token_true_pred_no_train[key] += 1

token_true_pred = sorted(token_true_pred.items(), key=lambda x: x[1], reverse=True)
token_true_pred_in_train = sorted(token_true_pred_in_train.items(), key=lambda x: x[1], reverse=True)
token_true_pred_no_train = sorted(token_true_pred_no_train.items(), key=lambda x: x[1], reverse=True)

token_true_pred
```
